```python
import json
import pandas as pd
```

```python
df = pd.read_csv('data/Resultados_Saber_11_nonull.csv')
df
```

```python
set(df['ESTU_DEPTO_RESIDE'].unique()) == \
set(df['COLE_DEPTO_UBICACION'].unique()) == \
set(df['ESTU_DEPTO_PRESENTACION'].unique())
```

```python
json_data = {}
depts = df['COLE_DEPTO_UBICACION'].unique()

for dept in depts:
	dept_df = df[df['COLE_DEPTO_UBICACION'] == dept]
	municipalities = dept_df['COLE_MCPIO_UBICACION'].unique()
	muni_dict = {}
	for municipality in municipalities:
		muni_df = dept_df[dept_df['COLE_MCPIO_UBICACION'] == municipality]
		schools = muni_df['COLE_NOMBRE_ESTABLECIMIENTO'].unique()
		muni_dict[municipality] = []
		for school in schools:
			school_df = muni_df[muni_df['COLE_NOMBRE_ESTABLECIMIENTO'] == school]
			school_dict = {school: {'bilingual': list(school_df['COLE_BILINGUE'].unique()),
				  		   'calendar': list(school_df['COLE_CALENDARIO'].unique()),
						   'journey': list(school_df['COLE_JORNADA'].unique()),
						   'nature': list(school_df['COLE_NATURALEZA'].unique()),
						   'character': list(school_df['COLE_CARACTER'].unique()),
						   'location_area': list(school_df['COLE_AREA_UBICACION'].unique())}}
			muni_dict[municipality].append(school_dict)
	json_data[dept] = muni_dict

display(json_data)
```

```python
with open('static/data_saber11.json', 'w') as json_file:
	json.dump(json_data, json_file, indent=4)
```
