```python
import json
import pandas as pd
```

```python
df = pd.read_csv('data/Resultados_Saber_Pro_nonull.csv')
df
```

```python
json_data = {}
depts = df['ESTU_INST_DEPARTAMENTO'].unique()

for dept in depts:
	dept_df = df[df['ESTU_INST_DEPARTAMENTO'] == dept]
	municipalities = dept_df['ESTU_INST_MUNICIPIO'].unique()
	muni_dict = {}
	for municipality in municipalities:
		muni_df = dept_df[dept_df['ESTU_INST_MUNICIPIO'] == municipality]
		insts = muni_df['INST_NOMBRE_INSTITUCION'].unique()
		muni_dict[municipality] = []
		for inst in insts:
			inst_df = muni_df[muni_df['INST_NOMBRE_INSTITUCION'] == inst]
			inst_dict = {inst: {
				'institution type': list(inst_df['INST_ORIGEN'].unique()),
				'instution character': list(inst_df['INST_CARACTER_ACADEMICO'].unique())
			}}
			muni_dict[municipality].append(inst_dict)
	json_data[dept] = muni_dict

display(json_data)
```

```python
with open('static/data_saber_pro.json', 'w') as json_file:
	json.dump(json_data, json_file, indent=4)
```
